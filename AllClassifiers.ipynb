## Implementation 2: RANSAC Regression (ignores the outliers)

```python
from sklearn.linear_model import RANSACRegressor

ransac = RANSACRegressor(LinearRegression(), 
                         max_trials=100, 
                         min_samples=50, 
                         loss='absolute_loss', 
                         residual_threshold=5.0, 
                         random_state=0)

ransac.fit(X, y)

inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)

line_X = np.arange(3, 31, 1)
line_y_ransac = ransac.predict(line_X[:, np.newaxis])
plt.scatter(X[inlier_mask], y[inlier_mask],
            c='steelblue', edgecolor='white', 
            marker='o', label='Inliers')
plt.scatter(X[outlier_mask], y[outlier_mask],
            c='limegreen', edgecolor='white', 
            marker='s', label='Outliers')
plt.plot(line_X, line_y_ransac, color='black', lw=2)   
plt.xlabel('Service time')
plt.ylabel('Age')
plt.legend(loc='upper left')
```

#### From the above plots of linear regression, we observe that the outliers do affect the model in a negative way - they tend to pull the line towards them. The simple linear regression (gradient descent) and Sklearn's linear regression generate the same regression lines. 

```python
#With test and train splitting
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=3)
```

#### Using attributes "Distance from Residence to Work vs Absenteeism time in hours" we will try to cluster people based on the number of pets they have.

```python
columns = ['Distance from Residence to Work', 'Absenteeism time in hours']
sns.pairplot(absentee_data.dropna(), hue='Pet', vars = columns)
```

```python
X = pet_filtered
X = X.iloc[:, [6, 20]].values
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=2, 
            init='random', 
            n_init=10, 
            max_iter=10,
            tol=1e-04,
            random_state=0)
y_km = km.fit_predict(X)

y_km
```

```python
km.cluster_centers_
```

```python
plt.scatter(X[y_km == 0, 0],
            X[y_km == 0, 1],
            s=50,
            c='lightgreen',
            marker='s',
            label='cluster 1')
plt.scatter(X[y_km == 1, 0],
            X[y_km == 1, 1],
            s=50,
            c='orange',
            marker='o',
            label='cluster 2')
plt.scatter(km.cluster_centers_[:, 0],
            km.cluster_centers_[:, 1],
            s=250,
            marker='*',
            c='red',
            label='centroids')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

### We find the optimal K value using the Elbow plot

```python
print('Distortion: %.2f' % km.inertia_)
```

```python
distortions = []
for i in range(1, 11):
    km = KMeans(n_clusters=i, 
                init='k-means++', 
                n_init=10, 
                max_iter=300, 
                random_state=0)
    km.fit(X)
    distortions.append(km.inertia_)
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.tight_layout()
plt.show()
```

### Finding the optimal K value using the Silhoutte coefficient

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

mini =100

range_n_clusters = list(range(2,10))
print('Clusters numbers in order from 2 to 10: \n', range_n_clusters)

for n_clusters in range_n_clusters: 
    clusterer = KMeans(n_clusters=n_clusters).fit(X)
    preds = clusterer.predict(X)
    centers = clusterer.cluster_centers_

    score = silhouette_score(X, preds, metric='euclidean')
    print("For n_clusters = {}, silhouette score is : {})".format(n_clusters, score))
    
    if(score< mini):
        p = n_clusters
        mini = score
    
print("The best Cluster is "+str(p)+" with score of"+str(mini))
```

#### Based on the initial position of cluster centres, it will suggest different silhoutte scores. Thus, the best cluster prediction itself will change after each run.

#### We can see that the Silhoutte coefficient and the Elbow method give two different values for K. However, giving higher preference to the Elbow method, and hence select K = ___ as the optimal value.
